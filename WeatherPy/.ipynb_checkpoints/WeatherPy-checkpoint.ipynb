```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import csv
from datetime import date

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_path = os.path.join("..","Output_Files","cities.csv")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Current Date Values
curdate = date.today().strftime("%m/%d/%Y")
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# # Print the city count to confirm sufficient count
print(f"Weather of {len(cities)} will be analyzed.")
```

```python
#Create Data frame to store API Weather Calls for each city
city_df = pd.DataFrame(cities,columns =['City'])

city_df["Cloudiness"] = ""
city_df["Country"] = ""
city_df["Date"] = ""
city_df["Humidity"] = ""
city_df["Latitude"] = ""
city_df["Longitude"] = ""
city_df["Max Temp"] = ""
city_df["Wind Speed"] = ""
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#API Calls
# Save config information and parameters

url = "http://api.openweathermap.org/data/2.5/weather?"

params = {
    "appid": api_key,
    "units": "imperial"
}


# use iterrows to iterate through pandas DataFrame
for index, row in city_df.iterrows():

    # use city name from city list to build URL for each city request
    # add a new key to our params dictionary that will store our keyword to search for
    city_parm = row["City"]
    params["q"] = city_parm

    #Request weather information form Open Weather Map
    response = requests.get(url, params).json()

    # Use exception handling in case any of our cities attributes are missing
    try:
        citycode=response["cod"]       
        print(f"Weather retrived for city: {city_parm}  Code:  {citycode}")
            
        city_df.loc[index, "Cloudiness"] = response["clouds"]["all"]
        city_df.loc[index, "Country"] = response["sys"]["country"]
        city_df.loc[index, "Date"] = response["dt"]
        city_df.loc[index, "Humidity"] = response["main"]["humidity"]
        city_df.loc[index, "Latitude"] = response["coord"]["lat"]
        city_df.loc[index, "Longitude"] = response["coord"]["lon"]
        city_df.loc[index, "Max Temp"] = response["main"]["temp_max"]
        city_df.loc[index, "Wind Speed"] = response["wind"]["speed"] 
                    
    except (KeyError, IndexError):
          print(f"Missing City: {city_parm}/result... skipping.")

print("------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create new df excluding the cities not found
city_df = city_df[city_df["Latitude"] != ""]

#Save cities data fram to csv file
city_df.to_csv (output_path, index = None, header=True)
```

#### Latitude vs. Temperature Plot

```python
# Max Tempurature Plot Chart

x_axis = city_df["Latitude"].tolist()
y_axis = city_df["Max Temp"].tolist()

fig= plt.figure(figsize=(8,6))

plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs. Max Temperature " + curdate)
plt.ylim(min(y_axis)*1.2, max(y_axis)*1.2)
plt.xlim(min(x_axis)*1.2, max(x_axis)*1.2)
plt.grid()

plt.scatter(x=x_axis, y=y_axis, marker="o", facecolors="blue", edgecolors="black", alpha=0.8)

# Save the Figure
plt.savefig("../Output_Files/CityLatitudeVSMaxTemp.png")
```
